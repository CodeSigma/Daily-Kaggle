```python
import re
splitted = []
for idx, row in rules.iterrows():
    splitted.append(re.split('>',row["rule"]))
    
```

```python
productids = pd.unique(data["itemid"])
# productids = list(map(str, productids))
print(len(productids))
# products = dict(zip(productids, []))
```

```python
products = dict()
for idx, row in data.iterrows():
    if row["itemid"] in products:
        products[row["itemid"]].append(row["orderid"])
    else:
        products[row["itemid"]] = [row["orderid"]]
```

```python
products
```

```python
def ins(lst1, lst2): 
    return list(set(lst1) & set(lst2))
```

```python
import math
values = []
for row in splitted_2:
    if len(row[0]) == 1:
        if len(row[1]) == 1:
            print(int(row[0][0]), int(row[1][0]))
            intercept = ins(products[int(row[0][0])], products[int(row[1][0])])
            
        else:
            print(int(row[0][0]), row[1])
            intercept = products[int(row[0][0])]
            for x in row[1]:
                intercept = ins(intercept, products[int(x)])
#             print(len(intercept)/len(row[0][0]))
        val = len(intercept)/len(products[int(row[0][0])])
        
    else:
        print(row[0], int(row[1][0]))
        intercept = products[int(row[1][0])]
        for x in row[0]:
            intercept = ins(intercept, products[int(x)])
#         print(len(intercept)/len(intercept(products[int(row[0][0])],products[int(row[0][1])] )))
        val = len(intercept)/len(ins(products[int(row[0][0])],products[int(row[0][1])] ))
    val = (math.floor(val*1000))
    print(val)
    values.append(val)
```
